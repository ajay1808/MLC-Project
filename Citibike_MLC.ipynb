```python
import requests
import sys
import pandas as pd
import time
from datetime import date, timedelta
import socket
import warnings
warnings.filterwarnings("ignore")
import datetime as dt
```

```python
jan_18=pd.read_csv("jan18.csv")
jan_18.head()
```

```python
jan_18['date'] = pd.to_datetime(jan_18['starttime']).dt.date
jan_18['start_hour'] = pd.to_datetime(jan_18['starttime']).dt.hour

jan_18['date'] = pd.to_datetime(jan_18['stoptime']).dt.date
jan_18['stop_hour'] = pd.to_datetime(jan_18['stoptime']).dt.hour

jan_18.head()
```

```python
jan_18['hour'] = jan_18[['start_hour','stop_hour']].mean(axis=1)
jan_18.head()
```

```python
def cleaning(df):
    df['date'] = pd.to_datetime(df['starttime']).dt.date
    df['start_hour'] = pd.to_datetime(df['starttime']).dt.hour
    df['stop_hour'] = pd.to_datetime(df['stoptime']).dt.hour
    df['hour'] = df[['start_hour','stop_hour']].mean(axis=1)
    
    bins = [0,4,8,12,16,20,24]
    df['HOUR_BIN'] = pd.cut(df['hour'], bins,labels=["0","4","8","12","16","20"])
    citi=df.groupby(['date','HOUR_BIN']).count()
    
    citi.rename({'tripduration': 'trip_count'}, axis=1, inplace=True)
    citi.drop(['hour','start_hour','stop_hour','start station id','start station latitude', 'start station longitude',
               'end station id','end station latitude', 'end station longitude', 'starttime', 'stoptime',
               'start station name', 'end station name', 'usertype','bikeid','birth year','gender'], 
            axis=1, inplace=True)
    citi.reset_index(inplace=True)
    return citi
```

```python
df=cleaning(jan_18)
df.columns
```

```python
df.head()
```

```python
year_2018_2019=['jan18.csv','feb18.csv','mar18.csv','apr18.csv','may18.csv','jun18.csv','jul18.csv','aug18.csv',
           'sep18.csv', 'oct18.csv','nov18.csv','dec18.csv', 
           'jan19.csv','feb19.csv','mar19.csv','apr19.csv','may19.csv','jun19.csv','jul19.csv','aug19.csv',
           'sep19.csv', 'oct19.csv','nov19.csv','dec19.csv']
citibike_18_19=pd.DataFrame(columns=df.columns)

for i in year_2018_2019:
    df=pd.read_csv(i)
    df=cleaning(df)
    citibike_18_19=citibike_18_19.append(df, ignore_index = True)

citibike_18_19
```
