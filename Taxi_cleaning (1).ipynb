```python
import pandas as pd
import geopandas as gpd
import urllib.request
import os
import numpy as np
from scipy import stats
import matplotlib.pyplot as plt
import warnings
warnings.filterwarnings('ignore')
```

```python
if os.path.exists('green_taxi/'):
    pass
else:
    os.makedirs('green_taxi')
list_ = []
for year in [2018,2019]:
    for month in range(1,13):
        if os.path.exists('green_taxi'+str(year)+str(month)+'/'):
            pass
        else:
            os.makedirs('green_taxi'+str(year)+str(month)+'/')
        urllib.request.urlretrieve ("https://d37ci6vzurychx.cloudfront.net/trip-data/green_tripdata_%s-%s.parquet"%(str(year),str(month).rjust(2, '0')), 
                                    'green_taxi'+str(year)+str(month)+'/'+"green_tripdata_%s-%s.parquet"%(str(year),str(month).rjust(2, '0')))
        data3 = pd.read_parquet("green_taxi/")
        data3 = data3[['lpep_pickup_datetime','lpep_dropoff_datetime','passenger_count','PULocationID','DOLocationID','fare_amount']]
    list_.append(data3)
```

```python
if os.path.exists('green_taxi_2019/'):
    pass
else:
    os.makedirs('green_taxi_2019/')

for year in [2019]:
    for month in range(1,13):
        urllib.request.urlretrieve ("https://d37ci6vzurychx.cloudfront.net/trip-data/green_tripdata_%s-%s.parquet"%(str(year),str(month).rjust(2, '0')), 
                                    "green_taxi_2019/green_tripdata_%s-%s.parquet"%(str(year),str(month).rjust(2, '0')))
        
```

```python
data_green_3 = pd.read_parquet("green_taxi_2019/")
data_green_3.head()
```

```python
data_green_3.shape
```

```python
data_green = pd.concat([data_green_2018, data_green_3], axis = 0)
data_green.shape
```

```python
validind = (data_green.passenger_count> 0) & (data_green.trip_distance > 0) & (data_green.fare_amount > 0)
data_green = data_green[validind]
data_green['lpep_pickup_datetime'] = pd.to_datetime(data_green['lpep_pickup_datetime'], format='%Y-%m-%d %H:%M:%S')
data_green=data_green.loc[data_green['passenger_count']<=7]
data_green['Date'] = data_green['lpep_pickup_datetime'].dt.date
data_green['Month'] = data_green['lpep_pickup_datetime'].dt.month
data_green['Year'] = data_green['lpep_pickup_datetime'].dt.year
data_green['DOW'] = data_green['lpep_pickup_datetime'].dt.dayofweek
data_green['Hour'] = data_green['lpep_pickup_datetime'].dt.hour
data_green = data_green.loc[data_green['Year'].isin([2018,2019])]
data_green= data_green[['Date','Hour','passenger_count']]
data_green.head()
```

```python
def binning(x,bins=6):
    return x-(x%(24//bins))

data_green['HOUR_BIN'] = data_green['Hour'].apply(binning)
data_green['HOUR_BIN'].unique()
```

```python
data_green = data_green[["Date","HOUR_BIN","passenger_count"]]
data_green.shape
```
